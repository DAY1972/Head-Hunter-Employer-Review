```python
import pandas as pd
import codecs
import random
```

```python
def load_txt_and_csv_line_by_line_input_path(path_to_load):
    '''it loads txt and csv files line by line'''
    path = path_to_load
    with codecs.open(path, "r") as file:
        f = file.read().split(' ') 
    return(f)
def load_txt_and_csv_line_by_line_input_path1(path_to_load):
    '''it loads txt and csv files line by line'''
    path = path_to_load
    with codecs.open(path, "r", encoding='utf-8') as file:
        f = file.read().splitlines()
    return(f)
```

```python
path_to_folder = r"C:\Users\Dima\Desktop\СберУниверситет\Дипломный проект\Диплом файлы"
proffes = 'список профессий.txt'
woina_i_mir = 'voyna-i-mir.txt'
prest_i_nak = 'prestuplenie-i-nakazanie.txt'
```

```python
w_i_m = load_txt_and_csv_line_by_line_input_path(path_to_load=path_to_folder + '//' + woina_i_mir)
prof = load_txt_and_csv_line_by_line_input_path1(path_to_load=path_to_folder + '//' + proffes)
p_i_n = load_txt_and_csv_line_by_line_input_path(path_to_load=path_to_folder + '//' + prest_i_nak)
```

```python
# загрузка реального файла (уже заменен на синтетический)
path_to_folder = r"C:\Users\Dima\Desktop\СберУниверситет\Дипломный проект\Диплом файлы"
file_to_load = "HeadHunter_test_dummy.csv" # "HeadHunter_train_dummy.csv"
path = path_to_folder + "//" + file_to_load

file_df = pd.read_csv(path)
file_df
```

```python
position = []
positive = []
negative = []

for i, p in enumerate(file_df['position']):
    
    random_integer = random.randint(0, len(prof) - 1)
    position.append(prof[random_integer])
    
    random_integer_w_i_m = random.randint(0, len(w_i_m) - 100)
    random_integer_p_i_n = random.randint(0, len(p_i_n) - 100)
    random_integer_text_w_i_m = random.randint(10, 40)
    random_integer_text_p_i_n = random.randint(10, 40)
    positive.append(' '.join(w_i_m[random_integer_w_i_m : random_integer_w_i_m + random_integer_text_w_i_m]))
    negative.append(' '.join(p_i_n[random_integer_p_i_n : random_integer_p_i_n + random_integer_text_p_i_n]))

file_df['position'] = position
file_df['positive'] = positive
file_df['negative'] = negative
```

```python
# полученный синтетический файл
file_df
```

```python
# сохраняем синтетический файл
path_to_folder = r"C:\Users\Dima\Desktop\СберУниверситет\Дипломный проект\Диплом файлы"
file_to_load = "HeadHunter_test_dummy.csv" # "HeadHunter_train_dummy.csv"
path = path_to_folder + "//" + file_to_load

file_df.to_csv(path_or_buf=path, index=False)
```
