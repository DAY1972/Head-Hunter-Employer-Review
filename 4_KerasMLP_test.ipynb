```python
import numpy as np
import pandas as pd
# Keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.optimizers import SGD, Adam, Adadelta, RMSprop
from keras import regularizers
import keras.backend as K
# Scaling data
from tensorflow.keras.models import load_model
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# load dataset
path_to_load = r'C:\Users\Dima\Desktop\СберУниверситет\Дипломный проект\Диплом файлы\file_df_set_test_MLP_new.csv'
df = pd.read_csv(path_to_load)
```

```python
df
```

```python
# загружаем векторайзер
pkl_filename = r'C:\Users\Dima\Desktop\СберУниверситет\Дипломный проект\Диплом файлы\TfidfVectorizerMLP.pkl'
with open(pkl_filename, 'rb') as file:
    vectorizer = pickle.load(file)

X = vectorizer.transform(df['text']).toarray()
```

```python
X.shape
```

```python
# загружаем модель, получаем вектор вероятностей классов
model_load_path = r'C:\Users\Dima\Desktop\СберУниверситет\Дипломный проект\Диплом файлы\KerasModel'
model = load_model(model_load_path)
y_pred = model.predict(X)
y_pred[0]
```

```python
# переводим вероятности в формат для сабмита
target = []
for i in range(y_pred.shape[0]):
    target.append('')
    for j in range(y_pred.shape[1]):
        if y_pred[i, j] > 0.2: # уровень отсечки для класса
            if len(target[i]) == 0:
                target[i] += str(j)
            else:
                target[i] += ',' + str(j)
df['target'] = target
target
```

```python
# сохраняем сабмит
path_to_save = r'C:\Users\Dima\Desktop\СберУниверситет\Дипломный проект\Диплом файлы\submit_MLP.csv'
df.to_csv(path_or_buf=path_to_save, columns=['review_id', 'target'], index=False)
```
